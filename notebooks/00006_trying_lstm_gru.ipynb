```python
from recibrew.core_nn import GRUBahdanauLightning
```

```python
gru_params = dict(
    embedding_dim=100, 
    hidden_dim=100, 
    enc_bidirectional=False, 
    enc_gru_layers=1,
    dropout=0.1
)
```

```python
gbl = GRUBahdanauLightning(gru_params=gru_params)
```

```python
from pytorch_lightning import Trainer, seed_everything
from pytorch_lightning.callbacks import EarlyStopping, ModelCheckpoint

seed_everything(888)
```

```python
import os

early_stop_callback = EarlyStopping(
   monitor='val_loss',
   min_delta=0.00,
   patience=3,
   verbose=False,
   mode='auto'
)

tmpdir = 'out_cp/'
ckpt = ModelCheckpoint(os.path.join(tmpdir, '{epoch}-{val_loss:.2f}'), save_top_k=2)

trainer = Trainer(early_stop_callback=early_stop_callback, checkpoint_callback=ckpt, 
                  accumulate_grad_batches=2, 
                  auto_lr_find=False, gpus=[0])
```

```python
trainer.fit(gbl)
```
